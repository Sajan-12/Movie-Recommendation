```python
#import relevant Library
import numpy as np
import pandas as pd
import os
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
#TfidfVectorizer is used convert text datya into numerical value
from sklearn.metrics.pairwise import cosine_similarity
import warnings
warnings.filterwarnings('ignore')
```

```python
#load the movies data
os.chdir('C:\\Users\\Asus\\Downloads')
data=pd.read_csv('movies.csv')
display(data)
```

```python
# extract importanmt features for Recomondation
selected_features=['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
#check null value in selected features
print(data[selected_features].isna().sum())
```

```python
#fill the missing value by null string
data[selected_features]=data[selected_features].fillna('')
print(data[selected_features].isna().sum())
```

```python
# combine all selected featues
combined_features=data['genres']+' '+data['keywords']+' '+data['tagline']+' '+data['cast']+' '+data['director']
display(combined_features)
```

```python
#Converting the text data to feature vectors
#This is to find cosine similarity 
#Vector shape is (4803, 17318). This is based on the number of distinct words.
#All the words will be converted to their equivalent numbers.

```

```python
vectorizer=TfidfVectorizer()
feature_vectors=vectorizer.fit_transform(combined_features)
feature_vectors.shape
```

```python
#getting cosine similarity
similarity_score=cosine_similarity(feature_vectors)
print(similarity_score)
```

```python
# take input movie name from user
movie_name=input('enter your favorite movie name: ')
```

```python
#Creating a list with all the movie names given in the dataset

list_of_all_movie_title=data['title'].tolist()
print(list_of_all_movie_title)
```

```python
print(len(list_of_all_movie_title))
```

```python
#Finding the close match  movie name given by the user
find_close_matches=difflib.get_close_matches(movie_name,list_of_all_movie_title)
print(find_close_matches)
```

```python
#display the closet match movie name
close_movie_name=find_close_matches[0]
print(close_movie_name)
```

```python
# find index of close match movie
index_of_the_movie=data[data.title==close_movie_name]['index'].values[0]
print(index_of_the_movie)
```

```python
#find similarity score of movie  based on index
similarity_score_of_user_movie=list(enumerate(similarity_score[index_of_the_movie]))
print(similarity_score_of_user_movie)
```

```python
#sort similarity score
soreted_similar_movies_score=sorted(similarity_score_of_user_movie,key=lambda x:x[1],reverse=True)
print(soreted_similar_movies_score)
```

```python
#find similar movie names
i=1
for movie in soreted_similar_movies_score:
    index=movie[0]
    name_from_index=data[data.index==index]['title'].values[0]
    if(i<21):
        print(i,'.', name_from_index)
        i=i+1;
    
```

```python
#Consolidated code 
movie_name=input('enter your favorite movie name:')
list_of_all_movie_title=data['title'].tolist()
print(list_of_all_movie_title)
print(len(list_of_all_movie_title))
find_close_matches=difflib.get_close_matches(movie_name,list_of_all_movie_title)
print(find_close_matches)
close_movie_name=find_close_matches[0]
print(close_movie_name)
index_of_the_movie=data[data.title==close_movie_name]['index'].values[0]
similarity_score_of_user_movie=list(enumerate(similarity_score[index_of_the_movie]))
print(similarity_score_of_user_movie)
soreted_similar_movies_score=sorted(similarity_score_of_user_movie,key=lambda x:x[1],reverse=True)
print(soreted_similar_movies_score)
```

```python
i=1
for movie in soreted_similar_movies_score:
    index=movie[0]
    name_from_index=data[data.index==index]['title'].values[0]
    if(i<21):
        print(i,'.', name_from_index)
        i=i+1;
```
